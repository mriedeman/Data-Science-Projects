```python
def get_data(start_date, end_date, savant_ids):  



    dictionary_list = []
    for i in savant_ids:   
        player_sabermetrics = statcast_pitcher(start_date, end_date, player_id = i)

        dictionary_list.append(player_sabermetrics)


    train_df = pd.concat(dictionary_list)

    train_df.to_csv('../data/pitching_masterfile10.csv',index = False)


    end_time = time.time()
    time_elapsed = end_time - start_time
    print(time_elapsed)
    return train_df
    
get_data('2018-04-01','2021-06-15',pitcher_ids)
```
